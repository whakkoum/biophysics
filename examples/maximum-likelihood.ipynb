```python
import numpy as np
import matplotlib.pyplot as plt

from scipy.special import beta
```

```python
l = 6
N = 10
B = 1/beta(l+1, N-l + 1)
```

```python
xi = np.linspace(0,1,1001)
P = B * xi**l * (1-xi)**(N-l)

plt.plot(xi, P)
```
